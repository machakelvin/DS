```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/kaggle/input/mobile-price-prediction/Cellphone.csv')

# Select the input features and target variable
features = ['Sale', 'weight', 'resoloution', 'ppi', 'cpu core', 'cpu freq', 'internal mem', 'Front_Cam']
target = 'Price'

# Normalize the input features
data[features] = (data[features] - data[features].mean()) / data[features].std()

# Add a column of ones for the bias term
data['bias'] = 1

# Create the feature matrix (X) and the target vector (y)
X = data[['bias'] + features].values
y = data[target].values

# Define the hyperparameters
learning_rate = 0.01
num_iterations = 1000

# Initialize the parameters
num_features = X.shape[1]
theta = np.zeros(num_features)

# Define the cost function (mean squared error)
def compute_cost(X, y, theta):
    m = len(y)
    predictions = np.dot(X, theta)
    cost = np.sum((predictions - y) ** 2) / (2 * m)
    return cost

# Perform gradient descent
cost_history = []

for i in range(num_iterations):
    predictions = np.dot(X, theta)
    errors = predictions - y
    gradient = np.dot(X.T, errors) / len(y)
    theta -= learning_rate * gradient
    
    cost = compute_cost(X, y, theta)
    cost_history.append(cost)

    if i % 100 == 0:
        print(f'Iteration {i}: Cost = {cost}')

# Plot the cost history
plt.plot(range(num_iterations), cost_history)
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Cost vs. Iteration')
plt.show()
```
